```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
def unique_values(columns):
    for column in columns:
        print(f"{column} : {data[column].unique()}")
```

```python
unique_values(data.columns)
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x=data['habitat'], data=data, hue='class', palette='dark')
plt.xlabel('Habitat')
plt.ylabel('Number of Mushroom')
plt.title('Poisionous and Palatable distribution of different habitat')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.stripplot(x=data['population'], y=data['habitat'], marker='o', color='m', size=6, edgecolor='white',orient='v')
plt.xlabel('Population')
plt.ylabel('Habitat')
plt.title('')
plt.legend()
plt.show()
```
